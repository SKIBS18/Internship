```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
rank = []
for i in soup.find_all('span',class_="rankings-table__pos-number"):
    rank.append(i.text.replace('\n',''))
    
rank
```

```python
rank.insert(0,1)
```

```python
rank
```

```python
rank_final = rank[0:10]
```

```python
rank_final
```

```python
name = []
for i in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
    name.append(i.text.replace('\n',''))
    
name
```

```python
name.insert(0,'Babar Azam')
```

```python
name
```

```python
name_final = name[0:10]
```

```python
name_final
```

```python
country = []
for i in soup.find_all('span',class_="table-body__logo-text"):
    country.append(i.text.replace('\n',''))
    
country
```

```python
country.insert(0,'PAK')
```

```python
country
```

```python
country_final = country[0:10]
```

```python
country_final
```

```python
rating = []
for i in soup.find_all('td',class_="table-body__cell rating"):
    rating.append(i.text.replace('\n',''))
    
rating
```

```python
rating.insert(0,892)
```

```python
rating
```

```python
rating_final = rating[0:10]
rating_final
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Ranking':rank_final,'Name':name_final,'Country':country_final,'Rating':rating_final})
df
```
