```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
header_tag = []
for i in soup.find_all('span',class_="mw-headline"):
    header_tag.append(i.text)
    
header_tag    
    
    
```

Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame


```python
page = requests.get('https://www.imdb.com/list/ls091520106/')
page
```

```python
content = BeautifulSoup(page.content)
content
```

```python
Name = []
for i in content.find_all('h3',class_="lister-item-header"):
    Name.append(i.text.replace('\n',''))
    
Name
```

```python
star = []
for i in content.find_all('div',class_="ipl-rating-star small"):
    star.append(i.text.replace('\n',''))
star    
```

```python
print(len(Name),len(star))
```

```python
import pandas as pd
df = pd.DataFrame({'Movie_Name & Year':Name,'Rating':star})
df
```

Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) and make data frame

```python
web = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
web
```

```python
details = BeautifulSoup(web.content)
```

```python
details
```

```python
ind_movie = []
for i in details.find_all('td',class_="titleColumn"):
    ind_movie.append(i.text.replace('\n',""))
    
    
ind_movie
```

```python
pointer = []
for i in details.find_all('td',class_="ratingColumn imdbRating"):
    pointer.append(i.text.replace('\n',''))
pointer
```

```python
year = []
for i in details.find_all('span',class_="secondaryInfo"):
    year.append(i.text)
year
```

```python
print(len(ind_movie),len(pointer),len(year))
```

```python
import pandas as pd
```

```python
table =  pd.DataFrame({'Movie_Name':ind_movie,'Rating':pointer,'Release Year':year})
table
```

```python
table.drop(table.index[100:],0,inplace=True)
table
```
