```python
driver = webdriver.Chrome(r'C:\Users\shail\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url = 'https://www.olx.in/cars_c84?filter=make_eq_maruti-suzuki'
driver.get(url)
```

```python
link = []
url = driver.find_elements(By.XPATH,"//li[@class='_31j8e']/a")
for u in url:
    link.append(u.get_attribute('href'))
```

```python
link
```

```python
len(link)
```

```python
driver.get(link[0])
```

```python
brand = []
price = []
variant = []
fuel = []
owner = []
location = []
for m in link:
    driver.get(m) 
    sleep(8)
    try:
        Bname = driver.find_element(By.XPATH,"//div[@class='_35xN1']").text
        brand.append(Bname)
    except NoSuchElementException as e:
        brand.append('--')
        pass
    
    try:
        p = driver.find_element(By.XPATH,"//div[@class='_3FkyT']").text
        price.append(p)
    except NoSuchElementException as e:
        price.append('--')
        pass
    try:
        var = driver.find_element(By.XPATH,"//div[@class='_3tLee']").text
        variant.append(var)
    except NoSuchElementException as e:
        variant.append('--')
        pass
    try:
        fue = driver.find_element(By.XPATH,"//div[@class='aOxkz']").text
        fuel.append(fue)
    except NoSuchElementException as e:
        fuel.append('--')
        pass
    try:
        ow = driver.find_element(By.XPATH,"//div[@class='_1gasz']").text
        owner.append(ow)
    except NoSuchElementException as e:
        owner.append('--')
        pass
    try:
        loc = driver.find_element(By.XPATH,"//div[@class='ZGU9S']").text
        location.append(loc)
    except NoSuchElementException as e:
        location.append('--')
        pass
    
    sleep(3) 
```
