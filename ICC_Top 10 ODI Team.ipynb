```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
rank = []
for i in soup.find_all('td',class_="table-body__cell table-body__cell--position u-text-right"):
    rank.append(i.text)
    
rank
```

```python
rank.insert(0,1)
```

```python
rank
```

```python
rank_final = rank[0:10]
```

```python
rank_final
```

```python
team = []
for i in soup.find_all('span',class_="u-hide-phablet"):
    team.append(i.text.replace('\n',''))
    
team
```

```python
team_final = team[0:10]
```

```python
team_final
```

```python
matches = []
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    matches.append(i.text.replace('\n',''))
    
matches
```

```python
matches_final=matches[0::2]

matches_final
```

```python
matches_final.insert(0,12)
```

```python
matches_final
```

```python
m = matches_final[0:10]
```

```python
m
```

```python
points = []
for i in soup.find_all('td',class_="table-body__cell u-center-text"):
    points.append(i.text.replace('\n',''))
    
points
```

```python
points_final = points[1::2]
```
