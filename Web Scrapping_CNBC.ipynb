```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.cnbc.com/world/?region=world')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
headline = []
for i in soup.find_all('div',class_="LatestNews-container"):
    headline.append(i.find("a",class_="LatestNews-headline").text)
    
headline
```

```python
newsLink = []
for i in soup.find_all('div',class_="LatestNews-container"):
    newsLink.append(i.find("a",class_="LatestNews-headline").get("href"))
    
newsLink
```

```python
time = []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
    
time
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Headline':headline,'Time':time,'Link':newsLink})
df
```
