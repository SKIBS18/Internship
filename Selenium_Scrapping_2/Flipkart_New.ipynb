```python
link = []
for i in range(1):
    url = driver.find_elements(By.XPATH,"//div[@class='_2kHMtA']/a")
    for u in url:
        link.append(u.get_attribute('href'))

```

```python
RAM = []
for m in link:
    driver.get(m)
    try:

        read_more = driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')     

        read_more.click()

    except NoSuchElementException:

        print("Exception Occured. Moving to next page")
        
        pass
    
    try:

        ram = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[2]/td[2]/ul/li')                

        RAM.append(ram.text)

    except NoSuchElementException:

        RAM.append('-')
        
        pass
    
    sleep(4)
    
    
        
```

```python
ROM = []
for m in link:
    driver.get(m)
    try:
        read_more = driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')     
        read_more.click()
    except NoSuchElementException:
        print("Exception Occured. Moving to next page")
        pass
    
    try:
        rom = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][4]/table[1]/tbody/tr[1]/td[2]/ul/li')                
        ROM.append(rom.text)
    except NoSuchElementException:
        ROM.append('-')
        pass
    
    sleep(4)
    
```

```python
ph_n = []
color = []
for m in link:
    driver.get(m)
    try:
        read_more = driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')     
        read_more.click()
    except NoSuchElementException:
        print("Exception Occured. Moving to next page")
        pass
    
    try:
        name = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][1]/table[1]/tbody/tr[3]/td[2]/ul/li')                
        ph_n.append(name.text)
    except NoSuchElementException:
        ph_n.append('-')
        pass
    try:
        clr = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][1]/table[1]/tbody/tr[4]/td[2]/ul/li')                
        color.append(clr.text)
    except NoSuchElementException:
        color.append('-')
        pass
    sleep(4)
```

```python
pc = []
sc = []
d = []
b_c = []
price = []

for m in link:
    driver.get(m)
    
    try:
        amount = driver.find_element(By.XPATH,'//*[@id="container"]/div/div[3]/div[1]/div[2]/div[2]/div/div[3]/div[1]/div/div').text                
        price.append(amount)
    except NoSuchElementException:
        price.append('-')
        pass
    
    try:
        read_more = driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _1FH0tX"]')     
        read_more.click()
    except NoSuchElementException:
        print("Exception Occured. Moving to next page")
        pass
    
    try:
        p1 = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[1]/td[2]/ul/li')                
        pc.append(p1.text)
    except NoSuchElementException:
        pc.append('-')
        pass
    try:
        s1 = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][5]/table[1]/tbody/tr[2]/td[2]/ul/li')                
        sc.append(s1.text)
    except NoSuchElementException:
        sc.append('-')
        pass
    try:
        d1 = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][2]/table[1]/tbody/tr[1]/td[2]/ul/li')                
        d.append(d1.text)
    except NoSuchElementException:
        d.append('-')
        pass
    try:
        b1 = driver.find_element(By.XPATH,'//div[@class="_3k-BhJ"][7]/table[1]/tbody/tr/td[2]/ul/li')                
        b_c.append(b1.text)
    except NoSuchElementException:
        b_c.append('-')
        pass
    
    
    sleep(4)
```
