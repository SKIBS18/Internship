```python
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
driver = webdriver.Chrome(r'C:\Users\shail\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url = 'https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-08-24&to=2022-08-26&guests=2&HostelNumber=&page=1'
driver.get(url)
```

```python
link = []

for i in range(3):
    print('Scraping page',i+1)
    url = driver.find_elements(By.XPATH,"//h2[@class='title title-6']/a")
    for u in url:
        link.append(u.get_attribute('href'))
    sleep(3)
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
```

```python
name = []
for i in range(3):
    print('Scraping page',i+1)
    nam = driver.find_elements(By.XPATH,"//h2[@class='title title-6']")
    for n in nam:
        name.append(n.text)
    sleep(3)
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
```

```python
distance = []
for w in range(3):
    print('Scraping page',w+1)
    dis = driver.find_elements(By.XPATH,"//span[@class='description']")
    for d in dis:
        if d.text is None:
            distance.append('--')
        else:
            distance.append(d.text)     
    sleep(3)
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
```

```python
distance
```

```python
price = []
privates_form = []
dorms_form = []

for o in range(3):
    print('Scraping page',o+1)
    for p_form in driver.find_elements(By.XPATH,"(//div[@class='price-col'])"):
        price.append(p_form.text)
    
    for private in range(0,len(price),2):
        privates_form.append(price[private]) 
             
    for dorms in range(1,len(price),2):
        dorms_form.append(price[dorms])
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break
```

```python
facility = []

for s in range(3):
    print('Scraping page',s+1)
    all_details = driver.find_elements(By.XPATH,'//div[@class="property-card"]')

    for details in all_details:
        try:
            if len(details.find_elements(By.XPATH,".//div[@class='has-wifi']"))>0:
                fac = details.find_elements(By.XPATH,".//div[@class='has-wifi']")
                for m in fac:
                     facility.append(m.text)
            else:
                facility.append("NA")
            
               
        except:
            pass
        
    
    sleep(3)
    try:
        next_button = driver.find_element(By.XPATH,'//div[@class="pagination-item pagination-next"]')
        next_button.click()
    except:
        break

    
```

```python
rating = []
t_review = []
o_review = []

for m in link:
    driver.get(m)
    try:
        rat = driver.find_element(By.XPATH,'//div[@class="score orange big"]').text
        rating.append(rat)
    
        
    except NoSuchElementException as e:
        rating.append('--')
        pass
    
    try:
        rev = driver.find_element(By.XPATH,'//div[@class="reviews"]').text
        t_review.append(rev)
    
        
    except NoSuchElementException as e:
        t_review.append('--')
        pass
    
    try:
        orev = driver.find_element(By.XPATH,'//div[@class="keyword"]').text
        o_review.append(orev)
    
        
    except NoSuchElementException as e:
        o_review.append('--')
        pass
    
    sleep(4)
        
```

```python
des = []

for m in link:
    driver.get(m)
    try:
        de = driver.find_element(By.XPATH,'//div[@class="content collapse-content"]').text
        des.append(de)
    
        
    except NoSuchElementException as e:
        des.append('--')
        pass
    
    sleep(4)
```
