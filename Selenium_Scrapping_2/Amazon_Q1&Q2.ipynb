```python
link = []
for i in range(3):
    print('Scraping page',i+1)
    url = driver.find_elements(By.XPATH,"//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for u in url:
        link.append(u.get_attribute('href'))
    next_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')    
    next_button.click()
    sleep(5)
```

```python
brand = []
price = []
exchange = []
delivery = []
avail = []
for m in link:
    driver.get(m)
    sleep(9)
    Bname = driver.find_element(By.XPATH,'//*[@id="productTitle"]').text
    brand.append(Bname)
    try:
        p = driver.find_element(By.XPATH,'//*[@id="corePriceDisplay_desktop_feature_div"]/div[1]/span[2]/span[2]/span[2]').text
        price.append(p)
    except NoSuchElementException as e:
        price.append('--')
        pass
    try:
        r_e = driver.find_element(By.XPATH,'//*[@id="maxBuyBackDiscountSection"]/span/span').text
        exchange.append(r_e)
    except NoSuchElementException as e:
        exchange.append('--')
        pass
    try:
        ed = driver.find_element(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span[1]').text
        delivery.append(ed)
    except NoSuchElementException as e:
        delivery.append('--')
        pass
    try:
        av = driver.find_element(By.XPATH,'//*[@id="availability"]/span').text
        avail.append(av)
    except NoSuchElementException as e:
        avail.append('--')
    
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Brand':brand,'Price':price,'Exchange':exchange,'Delivery':delivery,'Availability':avail,'Link':link})
df
```
