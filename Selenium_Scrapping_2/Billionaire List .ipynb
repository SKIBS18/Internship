```python
rank = []
ran = driver.find_elements(By.XPATH,'//div[@class="rank"]')
for a in ran:
    rank.append(a.text)  
```

```python
rank
```

```python
name = []
nam = driver.find_elements(By.XPATH,'//div[@class="personName"]')
for b in nam:
    name.append(b.text)  
```

```python
name
```

```python
worth = []
wor = driver.find_elements(By.XPATH,'//div[@class="netWorth"]')
for c in wor:
    worth.append(c.text) 
```

```python
worth
```

```python
age = []
ag = driver.find_elements(By.XPATH,'//div[@class="age"]')
for d in ag:
    age.append(d.text) 
```

```python
age
```

```python
country = []
coun = driver.find_elements(By.XPATH,'//div[@class="countryOfCitizenship"]')
for e in coun:
    country.append(e.text)
```

```python
country
```

```python
source = []
sor = driver.find_elements(By.XPATH,'//div[@class="source-column"]')
for f in sor:
    source.append(f.text)
```
