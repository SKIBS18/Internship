```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name = []
for i in soup.find_all('td',class_="table-body__cell rankings-table__name name"):
    name.append(i.text.replace('\n',''))
    
name
```

```python
name.insert(0,'Trent Boult')
name
```

```python
name_final = name[0:10]
name_final
```

```python
team = []
for i in soup.find_all('span',class_="table-body__logo-text"):
    team.append(i.text.replace('\n',''))
    
team
```

```python
team.insert(0,'NZ')
team
```

```python
team_final = team[0:10]
team_final
```

```python
rating = []
for i in soup.find_all('td',class_="table-body__cell rating"):
    rating.append(i.text.replace('\n',''))
    
rating
```

```python
rating.insert(0,726)
```

```python
rating_final = rating[0:10]
rating_final
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Player Name':name_final,'Team':team_final,'Rating':rating_final})
df
```
