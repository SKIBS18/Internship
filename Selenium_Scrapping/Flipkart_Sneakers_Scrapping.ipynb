```python
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
from time import sleep
```

```python
driver = webdriver.Chrome(r'C:\Users\shail\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
url = 'https://www.flipkart.com/'
driver.get(url)
```

```python
search_sneakers = driver.find_element(By.CLASS_NAME,"_3704LK")
search_sneakers
```

```python
search_sneakers.send_keys("sneakers")
```

```python
search_btn = driver.find_element(By.XPATH,"/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search_btn
```

```python
search_btn.click()
```

```python
brand = []
for i in range(3):
    print('Scraping page',i+1)
    brands_name = driver.find_elements(By.XPATH,'//div[@class="_2B099V"]')
    for b in brands_name:
        brand.append(b.text)
    next_button = driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')    
    next_button.click()
    sleep(3)
```

```python
len(brand)
```
