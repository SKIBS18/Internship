```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
title = []
for i in soup.find_all('li',class_="sc-9zxyh7-1 sc-9zxyh7-2 exAXfr jQmQZp"):
    title.append(i.find("h2",class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR").text)
    
title
```

```python
author = []
for i in soup.find_all('span',class_="sc-1w3fpd7-0 pgLAT"):
    author.append(i.text)
    
author
```

```python
date = []
for i in soup.find_all('span',class_="sc-1thf9ly-2 bKddwo"):
    date.append(i.text)
    
date
```

```python
link = []
for i in soup.find_all('li',class_="sc-9zxyh7-1 sc-9zxyh7-2 exAXfr jQmQZp"):
    link.append(i.find("a",class_="sc-5smygv-0 nrDZj").get("href"))
    
link
```

```python
print(len(title),len(author),len(date),len(link))
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Title':title,'Author':author,'Published Date':date,'Link':link})
df
```
